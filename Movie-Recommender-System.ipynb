```python
movies = movies.merge(credits,on='title')
```

```python
movies = movies[['movie_id','title','release_date','genres','overview','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast

def convert3(obj):
    j = 0
    L = []
    for i in ast.literal_eval(obj):
        if j<3:
            L.append(i['name'])
            j+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i.get("job")=="Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies = movies[['movie_id','title','release_date','tags']]
```

```python
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies.head()
```

```python
movies['tags'][0]
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
```

### For Stemming
Stemming of data means ['Love','Loving','Loved'] is converted into ['Love','Love','Love'] (helps in resolving ambiguity)

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
print(ps.stem('Loving'))
print(ps.stem('Loved'))
print(ps.stem('Love'))
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

### Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

### Calculating Cosine Similarity for each vector with other vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
